```python
age_group = [(15,25),(25,35),(35,45),(45,55),(55,65)]
person = [350, 375, 120, 80, 90]

#mean 
#find mmid point of each class

mid = []
for a,b in age_group:
    mid.append((a + b) / 2)
#find mean
total_fx = 0
total_x = 0
for i in range(len(person)):
    total_fx += person[i] * mid[i]
    total_x += person[i]
mean = total_fx / total_x
print("mean =", round(mean,2))

#median
#find cf
cf = []
sum_f = 0
for f in person:
    sum_f += f
    cf.append(sum_f)
n = sum_f
half = n / 2

#find which class contain median
for i in range(len(cf)):
    if cf[i] >= half:
        L = age_group[i][0] #lower limit 
        f = person[i]
        c = cf[i-1] if i > 0 else 0
        h = age_group[0][1] - age_group[0][0]
        median = L + ((half - c) / f) * h
        break
print("median =", round(median, 2))

#find mode

max_index = person.index(max(person))
L = age_group[max_index][0]
f1 = person[max_index]
f0 = person[max_index - 1] if max_index > 0 else 0 
f2 = person[max_index + 1] if max_index < len(person)-1 else 0
h = age_group[0][1] - age_group[0][0]

mode = L + ((f1 - f0) / (2 * f1 - f0 - f2)) * h
print("mode =" , round(mode, 2))
        


```
